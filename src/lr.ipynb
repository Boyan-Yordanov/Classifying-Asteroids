```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings(action='ignore')

from data_preprocessing import *
```

```python
df = pd.read_csv('../Asteroid.csv')
df
```

```python
X_train, X_train_scaled, X_test, X_test_scaled, y_train, y_test = process_data(df)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model_scaled = LogisticRegression()
lr_model.fit(X_train,y_train)
lr_model_scaled.fit(X_train_scaled,y_train)
print_accuracy(lr_model, lr_model_scaled, X_train, X_train_scaled, X_test, X_test_scaled, y_train, y_test)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(lr_model, X_train, y_train, cv=4, scoring="accuracy")
print(score)
```

```python
print("Average Cross Validation Model Accuracy: {:.3f}".format(score.mean()))
```

```python
C = [10, 1, 0.1]
penalty = ['l1', 'l2']
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
params = dict(penalty = penalty, C = C, solver = solver)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_search = RandomizedSearchCV(lr_model, params, cv = 5, scoring = 'accuracy', n_jobs= -1)
random_search.fit(X_test, y_test.ravel())
```

```python
plot_tuning(random_search)
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(lr_model, params, cv = 5, scoring = 'accuracy', n_jobs= -1)
grid_search.fit(X_test, y_test.ravel())
```

```python
plot_tuning(grid_search)
```

```python
print("Randomised Search")
print("The Best Accuracy of the Model is: {:.3f}".format(random_search.best_score_))
print("With Parameters: ", random_search.best_params_)
print("-----" *5)
print("Grid Search")
print("The Best Accuracy of the Model is: {:.3f}".format(grid_search.best_score_))
print("With Parameters: ", grid_search.best_params_)
```

```python
values = list(grid_search.best_params_.values())
print(values)
best_lr_model= LogisticRegression(C = values[0], penalty = values[1], solver = values[2])
best_lr_model.fit(X_train,y_train)
score = cross_val_score(best_lr_model, X_train, y_train, cv=4, scoring="accuracy")
print(score)
```

```python
print("Average Cross Validation Model Accuracy: {:.3f}".format(score.mean()))
```

```python
from sklearn import metrics
metrics.plot_roc_curve(best_lr_model, X_test, y_test)
plt.title('ROC on Linear Regression') 
plt.show()
```

```python
y_pred = lr_model_scaled.predict(X_test_scaled)
from sklearn.metrics import confusion_matrix
cfm = confusion_matrix(y_test, y_pred)
```

```python
group_names = ["True Neg","False Pos","False Neg","True Pos"]
group_counts = ["{0:0.0f}".format(value) for value in cfm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cfm.flatten()/np.sum(cfm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
df = pd.DataFrame(cfm, ["Not Dangerous","Dangerous"], ["Not Dangerous","Dangerous"])
plt.title("Confusion Matrix of Linear Regression", fontsize =16)
sns.heatmap(df, annot=labels,  fmt="", cmap='Blues')
```
